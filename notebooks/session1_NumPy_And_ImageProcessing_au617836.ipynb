## Image Translation

Process of taking an image and translating it to another kind of image, usually by changing the shape or orientation. <br> Translation = shifting of the image along the x- and y-axes. We can do this by using matrix multiplication. 

```python
# Reloading the original image of the t-rex
path_to_image = os.path.join("..", "data", "img", "trex.png")
image = cv2.imread(path_to_image)
jimshow(image, "Original")
```

__Translating__

We are essentially creating a new matrix when translating it by multiplying the elements in the matrix. 
The tx number specifies how much we are going to move the image to the right/left
The ty number specifies how much we are going to move the image up/down.

__Shifting image down and right__

Translation matrix <br>

shift 25 pixels on x-axis and 50 pixels on the y-axis

```python
# Creating an array of floats. Each row is created individually and have their own individual square brackets.
# This is a multiplication matrix that we can multiply the matrix with which shifts the image. 
M = np.float64([[1, 0, 25],
                [0, 1, 50]])
# We are shifting the image 25 pixels along the x-axis and 50 pixels along the y-axis
```

```python
# Specify width and height of the image
width = image.shape[1]
height = image.shape[0]
```

```python
# We create a shifted image using an Affine transformation which preserves colineratiry (everything stays parallel and on the same line etc.)
# Keep in mind that when performing a translation of an image we use the cv2.warpAffine() function
shifted = cv2.warpAffine(image, M, (width, height))
```

```python
# Showing the image that we have shifted
jimshow(shifted, "Shifted down and right")
```

__Shifting up and left__

Shift 100 pixels left, and 100 pixels up

```python
# The point of reference is 0,0 which means that if we want to go up we need to use negative values to go up the array
M = np.float64([[1, 0, -100 ],
                [0, 1, -100]])
```

```python
shifted = cv2.warpAffine(image, M, (width, height))
```

```python
jimshow(shifted, "Shifted up and left")
```

__What does this do?__

```python
M = np.float64([[1, 0, 0],
                [0, 1, 100]])
shifted = cv2.warpAffine(image, M, (width, height))
jimshow(shifted, "Shifted down")
# It shifts it down
```

__Writing a function for images__

We want to create a function that allows us to change the shifting parameters in the multiplication matrix. 

```python
def translate(image, x, y):
    # Define translation matrix
    M = np.float64([[1, 0, x],
                  [0, 1, y]])
    
    # Perform translation on the chosen image
    shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
    
    # Return the translated image
    return shifted
```

^Remember that image.shape[1] is the columns (width of the image) and the image.shape[0] is the rows (heigth of the image)

```python
# Now let's try to use the function we just created
shifted_image = translate(image, 50, 75)
jimshow(shifted_image, "Shifted right and down")
```
